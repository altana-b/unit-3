# Assignment 6.7 — Fibonacci Sequence

Your challenge is to write a Python function big_fibonacci that takes a number n and returns the first number of the Fibonacci sequence that has n digits. For instance, 1 is the first number with 1 digit in the sequence, 13 is the first number with 2 digits in the sequence, and 144 is the first number with 3 digits in the sequence.

```python
def big_fibonacci(n):
    if n==1:
        return 1
    a=1
    b=1
    c=a+b
    while len(str(c))<n:
        a=b
        b=c
        c=a+b
    return c
#A return statement is used to end the execution of the function call and “returns” the result (value of the expression following the return keyword) to the caller. 
#The statements after the return statements are not executed.
```

```python
big_fibonacci(1)
```

```python
big_fibonacci(2)
```

```python
big_fibonacci(10)
```

# Assignment 6.8 — FizzBuzz

Write a Python script called fizzbuzz.py that prints something for each number n from 1 to 1000, according to the following rules:

* If n is divisible by 3, print "Fizz".
* If n is divisible by 5, print "Buzz".
* If n is divisible by both 3 and 5, print "FizzBuzz".
* If n is not divisible by either 3 or 5, just print the number n.
So, your program should print out 1000 lines of text in total, and the first 20 lines should look like this:

```python
n=0
while 0<=n<=999:
    n=n+1
    if n%3==0 and n%5==0:
        print ('FizzBuzz')
    elif n%3==0 and n%5!=0:
        print('Fizz')
    elif n%3!=0 and n%5==0:
        print('Buzz')
    else:
        print(n)
```

# Assignment 6.9 — Prime Number Test

```python
def is_prime(n):
    if n<=1:
        print('Number has to be greater than 1 to be prime!')
    else:
        i=1
        while i<n-1:
            i=i+1
            result=n%i
            if result==0:
                return False
        else:
            return True
```

```python
print(is_prime(2))
print(is_prime(6))
print(is_prime(8))
print(is_prime(17))
print(is_prime(23))
print(is_prime(25))
```

# Assignment 6.10 — Find a Lot of Prime Numbers

Write a Python function called primes_below that takes a number n and returns a list of all prime numbers less than n.

```python
def primes_below(n):
    output=[]
    i=1
    while i+1<n:
        i=i+1
        if is_prime(i)==True:
            output.append(i)
    return output
```

```python
primes_below(17)
```

```python
primes_below(8)
```

```python
primes_below(2)
```

# Assignment 6.11 — Find Palindrome Primes

A number (or word) is a palindrome if it looks the same forwards or backwards. For instance, the number 12321 is a palindrome, because if you reverse its digits you get the same number.
* Here are all of the three-digit primes which are also palindromes, packed in a Python list.
* [101, 131, 151, 181, 191, 313, 353, 373, 383, 727, 757, 787, 797, 919, 929]
* Your task is to create a Python program called palindrome_primes.py that creates a list of all of the five-digit prime numbers which are palindromes. Your program can do many steps, but it should only print the result list.

```python
output=[]
n=10000
while len(str(n))==5:
    n=n+1
    if is_prime(n)==True:
        if str(n)[0]==str(n)[4] and str(n)[1]==str(n)[3]:
            output.append(n)
print(output)
```

```python
#3 digit version
output=[]
n=100
while len(str(n))==3:
    n=n+1
    if is_prime(n)==True:
        if str(n)[0]==str(n)[2]:
            output.append(n)
print(output)
```

# Assignment 6.13 — Triangle Class

* Create a Triangle class. Provide three numbers, giving the length of each side. 
* Your rectangle class should have perimeter, area, and scale methods in analogy with the same methods on the Rectangle class.
* Additionally, add two more methods to the Triangle class. First, implement a is_valid method so that t.is_valid() returns True if t is a valid triangle, or False otherwise. For example, there is no triangle with side lengths 1, 3, and 5. 
* Next, implement a is_right method that returns True if the triangle is a right triangle (having a 90 degree angle) and False otherwise. 

```python
import math as m
```

```python
class Triangle():
    def __init__(self,a,b,c):
        self.length1=a
        self.length2=b
        self.length3=c

    def perimeter(self):
        return self.length1+self.length2+self.length3

    def area(self):
        return m.sqrt((0.5*self.perimeter())*((0.5*self.perimeter())-self.length1)*((0.5*self.perimeter())-self.length1)*((0.5*self.perimeter())-self.length1))
            
    def scale(self,scale_factor):
        return Triangle(scale_factor*self.length1, scale_factor*self.length2,scale_factor*self.length3)

    def is_valid(self):
        #the sum of 2 sides of a triangle must be greater than the third side
        if (self.length1<(self.length2+self.length3)) and (self.length2<(self.length1+self.length3)) and (self.length3<(self.length1+self.length2)):
            return True
        else:
            return False
        
    #def is_valid2(self):
        #if (self.length1>=(self.length2+self.length3)) or (self.length2>=(self.length1+self.length3)) or (self.length3>=(self.length1+self.length2)):
            #return False
        #else:
            #return True
    
    def is_right(self):
        #Pythagorean theorem
        if self.length1==m.sqrt((self.length2)**2+(self.length3)**2) or self.length2==m.sqrt((self.length1)**2+(self.length3)**2) or self.length3==m.sqrt((self.length1)**2+(self.length2)**2):
            return True
        else:
            return False
```
